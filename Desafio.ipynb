### Pré processamento
Pré processar os dados e passar para um Dict e realizar outros processos: tokenização, etc.

```python
from datasets import Dataset, DatasetDict
from sklearn.model_selection import train_test_split

def setDatasetConverter(data):
    """
    Esta função converte um conjunto de dados em um formato adequado para uso com a biblioteca Hugging Face.

    Parâmetros:
    - data (list): Uma lista de dicionários, onde cada dicionário contém campos 'id', 'tokens' e 'ner_tags'.

    Retorna:
    - dataset (DatasetDict): Um objeto DatasetDict do Hugging Face contendo os dados convertidos nas divisões "train" e "valid".
    """
    try:
        # Dividindo os dados em subconjuntos de treinamento e validação
        trainData, validData = train_test_split(data, test_size=0.2, random_state=42)

        # Inicializa as listas para cada campo
        train_ids, valid_ids = [], []
        train_tokens, valid_tokens = [], []
        train_ner_tags, valid_ner_tags = [], []

        # Itera sobre os dados de treinamento e validação para extrair cada campo
        for item in trainData:
            train_ids.append(item["id"])
            train_tokens.append(item["tokens"])
            train_ner_tags.append(item["ner_tags"])

        for item in validData:
            valid_ids.append(item["id"])
            valid_tokens.append(item["tokens"])
            valid_ner_tags.append(item["ner_tags"])

        # Cria o dicionário para o dataset do Hugging Face para treinamento e validação
        train_hfDataset = {
            "id": train_ids,
            "tokens": train_tokens,
            "ner_tags": train_ner_tags
        }

        valid_hfDataset = {
            "id": valid_ids,
            "tokens": valid_tokens,
            "ner_tags": valid_ner_tags
        }

        # Cria datasets do Hugging Face a partir dos dicionários
        train_dataset = Dataset.from_dict(train_hfDataset)
        valid_dataset = Dataset.from_dict(valid_hfDataset)

        # Cria um DatasetDict com as divisões "train" e "valid"
        dataset = DatasetDict({"train": train_dataset, "valid": valid_dataset})

        return dataset

    except Exception as e:
        print(f"Erro ao converter dados para o formato do Hugging Face: {str(e)}")
        raise e


```

```python
from transformers import AutoTokenizer
# Importar o Tokenizador
tokenizer = AutoTokenizer.from_pretrained("distilbert/distilbert-base-uncased")
```

```python
def tokenizeAndAlignLabels(examples):
    """
    Tokeniza e alinha rótulos para os exemplos fornecidos.

    Args:
        examples (dict): Um dicionário contendo os campos 'tokens' e 'ner_tags'.

    Returns:
        tokenizedInputs (dict): Um dicionário contendo as entradas tokenizadas com rótulos alinhados.
    """
    tokenizedInputs = tokenizer(examples["tokens"], truncation=True, is_split_into_words=True)

    labels = []
    for i, label in enumerate(examples["ner_tags"]):
        wordIds = tokenizedInputs.word_ids(batch_index=i)  # Mapeia tokens para suas respectivas palavras.
        previousWordIdx = None
        labelIds = []
        for wordIdx in wordIds:  # Define os tokens especiais como -100.
            if wordIdx is None:
                labelIds.append(-100)
            elif wordIdx != previousWordIdx:  # Rotula apenas o primeiro token de uma palavra dada.
                labelIds.append(label[wordIdx])
            else:
                labelIds.append(-100)
            previousWordIdx = wordIdx
        labels.append(labelIds)

    tokenizedInputs["labels"] = labels
    return tokenizedInputs
```

```python
# Essa é a lista dos labels do ner_tags do Dataset
labelList = ['NULO', 'CIDADE', 'VALOR', 'CPF']
```

```python
datasetTokenizado = dataset.map(tokenizeAndAlignLabels, batched=True)
```

```python
from transformers import DataCollatorForTokenClassification
# Cria o Collator para o Modelo
dataCollator = DataCollatorForTokenClassification(tokenizer=tokenizer)
```

```python
import evaluate
# Essa é avaliação
seqeval = evaluate.load("seqeval")
```

```python
import numpy as np

def computeMetrics(predictions_and_labels):
    """
    Calcula métricas de avaliação para predições e rótulos.

    Args:
        predictions_and_labels (tuple): Uma tupla contendo as predições e os rótulos.

    Returns:
        dict: Um dicionário contendo as métricas de avaliação calculadas.
    """
    predictions, labels = predictions_and_labels
    predictions = np.argmax(predictions, axis=2)

    true_predictions = [
        [labelList[p] for (p, l) in zip(prediction, label) if l != -100]
        for prediction, label in zip(predictions, labels)
    ]
    true_labels = [
        [labelList[l] for (p, l) in zip(prediction, label) if l != -100]
        for prediction, label in zip(predictions, labels)
    ]

    results = seqeval.compute(predictions=true_predictions, references=true_labels)
    return {
        "precision": results["overall_precision"],
        "recall": results["overall_recall"],
        "f1": results["overall_f1"],
        "accuracy": results["overall_accuracy"],
    }
```

```python
id2label = {
    0: "NULO",
    1: "CIDADE",
    2: "VALOR",
    3: "CPF",
}
label2id = {
    "NULO": 0,
    "CIDADE": 1,
    "VALOR": 2,
    "CPF": 3,
}
```

### Prepara o modelo e treinamento

```python
from transformers import AutoModelForTokenClassification, TrainingArguments, Trainer

model = AutoModelForTokenClassification.from_pretrained(
    "distilbert/distilbert-base-uncased", num_labels=4, id2label=id2label, label2id=label2id
)
```

```python
trainingArgs = TrainingArguments(
    output_dir="NER-challenge",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=15,
    weight_decay=0.01,
    evaluation_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    push_to_hub=True,
)

trainer = Trainer(
    model=model,
    args=trainingArgs,
    train_dataset=datasetTokenizado["train"],
    eval_dataset=datasetTokenizado["valid"],
    tokenizer=tokenizer,
    data_collator=dataCollator,
    compute_metrics=computeMetrics,
)

```

```python
# Treina o modelo
trainer.train()
```

```python
# Envia o modelo para o HF
trainer.push_to_hub()
```
